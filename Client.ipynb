# CLIENT

```python
import socket
import sys
import os
from IPython.display import clear_output

# Initialise socket stuff
TCP_IP = "127.0.0.1" # Only a local server
TCP_PORT = 8000 # Just a random choice
BUFFER_SIZE = 1024 # Standard choice
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

def conn():
    # Connect to the server
    print ("Sending server request...")
    try:
        s.connect((TCP_IP, TCP_PORT))
        print ("Connection sucessful")
    except:
        print ("Connection unsucessful. Make sure the server is online.")

def upld():
    s.send("UPLD".encode())

    f = open("contoh.txt", "rb")
    
    try:
        print("Read data")
        byte = f.read(BUFFER_SIZE)
        
        while byte != b'':
            print("Read data...")
            s.send(byte)
            byte = f.read(BUFFER_SIZE)
            
    finally:
        print("File berhasil dikirim")
        f.close()
        s.close()

def dwld():
    s.send("DWLD".encode())
    
    f = open("hasil_download.txt", "w+b")
    while 1:
#         print("Receive Data")
        RECEIVED = s.recv(BUFFER_SIZE)
        f.write(RECEIVED)

#         print("Receive Data 2")
        if RECEIVED == b'1':
            print("File telah di download")
            break
    
    f.close()
    s.close()

def quit():
    s.send("QUIT".encode())
    s.close()
    print ("Server connection ended")
    return

def most():
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    TCP_IP = "127.0.0.1" # Only a local server
    TCP_PORT = 8000 # Just a random choice
    s.connect((TCP_IP, TCP_PORT))
    
    s.send("MOST".encode())
    l = s.recv(BUFFER_SIZE)
    print(l.decode())
    
    s.close()

print ("\n\nWelcome to the FTP client.\n\nCall one of the following functions:\nCONN           : Connect to server\nUPLD file_path : Upload file\nDWLD file_path : Download file\nQUIT           : Exit")

while True:
#     print("INPUT")
    x = input("Input command: ")
#     print("test")
    if x == "UPLD":
        x.encode()
#         clear_output(wait=True)
        upld()
    elif x == "CONN":
        conn()
    elif x == "DWLD":
        x.encode()
#         clear_output(wait=True)
        dwld()
    elif x == "QUIT":
        x.encode()
        quit()
        break
    elif x == "MOST":
        x.encode()
        most()
    else:
        print ("Command not recognised; please try again")
        x
```
